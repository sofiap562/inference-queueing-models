## Service Time vs Batch Size relationhip

```bash
curl http://triton_server:8000/v2/models/food_classifier_onnx/versions/1/stats
```

Now I will run benchmarks for a constant request rate range 162:729:81 (with constant distribution) for various batch sizes {1,8,10,12,16,18,20,22,24,26,28,30,32} while the model's configuration has a set max batch number of 32.

```bash
perf_analyzer -u triton_server:8000  -m food_classifier_onnx -b 8 --shape IMAGE:3,224,224 \
    --request-distribution=constant --request-rate-range 162:648:81 --async -f const_b8.csv
```
