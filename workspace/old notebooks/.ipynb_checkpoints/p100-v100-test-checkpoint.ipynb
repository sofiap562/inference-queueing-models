## Server configuration:

* Max batch size is 8
* Configured
```
dynamic_batching {}
```

Note: to edit configuration,

```
vim summer2025/models/food_classifier_onnx/config.pbtxt
```

and to re-load after editing configuration,

```
docker compose -f ~/summer2025/docker/docker-compose-p100-v100.yaml build triton_server_p100 
docker compose -f ~/summer2025/docker/docker-compose-p100-v100.yaml up triton_server_p100 --force-recreate -d

docker compose -f ~/summer2025/docker/docker-compose-p100-v100.yaml build triton_server_v100 
docker compose -f ~/summer2025/docker/docker-compose-p100-v100.yaml up triton_server_v100 --force-recreate -d
```
